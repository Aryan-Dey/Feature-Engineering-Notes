```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E:\\Notes\\Statistics\\Feature Engineering\\Datasets\\titanic_toy.csv')
df.head()
```

Rows contains different types of Numerical & Categorical Data
- Cabin contains - C85, C123
- Ticket contains - A/5 21171, PC 17599
- number - 5,3,A,6

```python
# First Transforming Number column - 
df['number_num'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df['number_cat'] = np.where(df['number_num'].isnull(), df['number'], np.nan)

# Filling Missing Values
df['number_num'].fillna(0, inplace=True)
df['number_cat'].fillna('Missing', inplace=True)

df.head()
```

```python
# Transforming Cabin Column
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)')
df['cabin_cat'] = df['Cabin'].str.extract(r'([A-Za-z]+)')

df.head()
```

```python
# Transforming Ticket Column - A/5 21171, 113803
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1]) #Extracts the last nos.
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')


df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0]) #Extracts the first part
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat']) #If the rpw only has digit so cat mai nan se replace ho jayega

df.head(10)
```
